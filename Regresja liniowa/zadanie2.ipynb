```python
import numpy as np

import statsmodels.api as sm

import statsmodels.formula.api as smf

import pandas as pd
import plotnine as p9
```

```python
beauty=pd.read_csv("beauty.csv")    
```

```python
#Model predyktor - piekno (btystdave) przewidywanie - ocena kursu (courseevaluation)
results = smf.ols('courseevaluation ~ btystdave', data=beauty).fit()
wyn=results.params
results.summary()
#Dopasownaie modelu 0.034
```

```python
#Wizualizacja modelu
fig=(p9.ggplot(p9.aes(x='btystdave',y='courseevaluation'),data=beauty)
      +p9.geom_jitter(width=0.1)
      +p9.geom_abline(p9.aes(intercept=wyn['Intercept'],slope=wyn['btystdave'])))
fig
```

```python
#residua
df=beauty
df['cs_pred']=results.predict()
df['residuals']=df['courseevaluation']-df['cs_pred']
fig1_res=(p9.ggplot(p9.aes(x='btystdave',y='residuals'),data=beauty)
      +p9.geom_point())
fig1_res
```

```python
#Przeanalizowannie korelacji pomiędzy współczynnikami
pd.set_option('display.max_rows', 500)
beauty.corr()['courseevaluation']
#wysoka korelacja dla profevaluation
```

```python
#Model predyktor - piekno (btystdave), wiek (age) przewidywanie - ocena kursu (courseevaluation)
results = smf.ols('courseevaluation ~ btystdave + age ', data=beauty).fit()
wyn=results.params
results.summary()
#Dopasownaie modelu Adj.R-squared 0.032
```

```python
#Model predyktor - piekno (btystdave), płeć (female) przewidywanie - ocena kursu (courseevaluation)
results = smf.ols('courseevaluation ~ btystdave + female ', data=beauty).fit()
wyn=results.params
results.summary()
#Dopasownaie modelu Adj.R-squared 0.062
```

```python
#Model predyktor - piekno (btystdave), język kursu (nonenglish), płeć(female) 
#przewidywanie - ocena kursu (courseevaluation)
results = smf.ols('courseevaluation ~ btystdave + nonenglish + female', data=beauty).fit()
wyn=results.params
results.summary()
#Dopasownaie modelu Adj.R-squared 0.081
```

```python
#Model predyktor - ocena profesora (profevaluation)
#przewidywanie - ocena kursu (courseevaluation)
results = smf.ols('courseevaluation ~ profevaluation', data=beauty).fit()
wyn=results.params
results.summary()
#Dopasownaie modelu Adj.R-squared 0.874
```
